This notebook contains a quick hack to extract text from images. It may come in handy to extract text from the images in the AnTeDe slides.

```python
!pip install pytesseract
!apt-get install -y tesseract-ocr

import pytesseract
from PIL import Image
import requests
from io import BytesIO
from google.colab import files

def ocr_jpg(jpg_file):
    image = Image.open(jpg_file)
    text = pytesseract.image_to_string(image)
    return text


```

```python
print("Please upload your JPG file")

uploaded = files.upload()
file_name = list(uploaded.keys())[0]

text = ocr_jpg(BytesIO(uploaded[file_name]))
print("\nText extracted from the image:")
print(text)
```
